```python
# Q2. What is an object in Python?

# ANSWER

'''In Python, an object is a fundamental concept that represents a data structure that combines both data (attributes) and methods (functions) that operate on that data.'''

# FOR EXAMPLE;

class animal:
    def __init__(self, species, sound):
        self.species = species
        self.sound = sound
        
    def behaviour(self):
        print('{} makes {} sound'.format(self.species, self.sound))
        
DOG = animal('Doberman', 'Bhow Bhow')
DOG.behaviour()



# In above example, Animal is a class which as attribute such as sepcies and sound and method which is behaviour. Now in this case, DOG is an object which is combining both attribute and mehtod 
```

```python
#Q5. What is the difference between class variables and instance variables in Python?

# ANSWER -->

# In Python, both class variables and instance variables are used to store data within a class, but they serve different purposes and have different scopes.

# Class variable --> Class variables are variables that are shared among all instances (objects) of a class. They are defined at the class level and are the same for all instances of the class.
# Instance variable--> Instance variables are variables that are specific to each instance (object) of a class. They are defined within the methods of the class, usually within the __init__ method. Each instance has its own separate set of instance variables.





# FOR EXAMPLE;

class animal:
    
    Nature =  'Human Friendly' 
    
    def __init__(self, species, sound):
        self.species = species
        self.sound = sound
        
    def behaviour(self):
        print('{} makes {} sound'.format(self.species, self.sound))
        
DOG = animal('Doberman', 'Bhow Bhow')
CAT = animal('Persian', 'Meow')

print('When object is Dog')
print(DOG.sound)
print(DOG.species)
print(DOG.Nature) 

print('\n')

print('When object is Cat')
print(CAT.sound)
print(CAT.species)
print(CAT.Nature) 


# In both the cases, the instances are different but they are sharing the same class variable which is nature.
```

```python
# Q7. For a library management system, you have to design the "Book" class with OOP principles in mind. The “Book” class will have following attributes:
#a. title: Represents the title of the book.
#b. author: Represents the author(s) of the book.
#c. isbn: Represents the ISBN (International Standard Book Number) of the book.
#d. publication_year: Represents the year of publication of the book.
#e. available_copies: Represents the number of copies available for checkout. 

#The class will also include the following methods:
#a. check_out(self): Decrements the available copies by one if there are copies available for checkout.
#b. return_book(self): Increments the available copies by one when a book is returned.
# c. display_book_info(self): Displays the information about the book, including its attributes and the number of available copies.


class BOOK:
    def __init__(self, title, author, isbn, publication_year, available_copies):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.publication_year = publication_year
        self.available_copies = available_copies
        
    def check_out(self, quantity):
        self.available_copies -=quantity
        
    def return_book(self, quantity):
        self.available_copies +=quantity
        
    def display_book_info(self):
        print('The total number of available books are {}'.format(self.available_copies))


        
######################################################################################################################################################################################################



Book1 = BOOK('money matter', 'avinash',123, 2019, 10)

print("Number of total Books", Book1.available_copies)

Book1.check_out(5)
print("Number of Books after checkout ", Book1.available_copies)

Book1.return_book(2)
print("Number of Books after Reutrn ", Book1.available_copies)

Book1.display_book_info()

```

```python
# Q8. For a ticket booking system, you have to design the "Ticket" class with OOP principles in mind. The “Ticket” class should have the following attributes:
#a. ticket_id: Represents the unique identifier for the ticket.
#b. event_name: Represents the name of the event.
#c. event_date: Represents the date of the event.
#d. venue: Represents the venue of the event.
#e. seat_number: Represents the seat number associated with the ticket.
#f. price: Represents the price of the ticket.
#g. is_reserved: Represents the reservation status of the ticket. 

#The class also includes the following methods:
#a. reserve_ticket(self): Marks the ticket as reserved if it is not already reserved.
#b. cancel_reservation(self): Cancels the reservation of the ticket if it is already reserved.
#c. display_ticket_info(self): Displays the information about the ticket, including its attributes and reservation status.



class TICKET:
    def __init__(self, ticket_id, event_name, event_date, venue, seat_number, price, is_reserved):
        self.ticket_id = ticket_id
        self.event_name = event_name
        self.event_date = event_date
        self.venue = venue
        self.seat_number = seat_number
        self.price = price
        self.is_reserved = is_reserved
        
    def reserve_ticket(self, ticket_id1):
        
        if ticket_id1 in self.ticket_id and self.is_reserved =="PENDING":
            self.is_reserved  = 'RESERVED'
        
        elif ticket_id1 in self.ticket_id and self.is_reserved =="RESERVED":
            print('THIS TICKET IS ALREADY RESERVED')
        else:
            print('Invalid ticket, Please check the ticket number')

            
    def cancel_reservation(self, ticket_id1):
        if ticket_id1 in self.ticket_id and self.is_reserved =="RESERVED":
            self.is_reserved  = 'CANCELLED'
                
        elif ticket_id1 in self.ticket_id and self.is_reserved =="PENDING":
            self.is_reserved  = 'CANCELLED'
                
        else:
            print("Invalid ticket, Please check the ticket number")
    
    def display_ticket_info(self):
        if  self.is_reserved  == 'RESERVED':
            print('Your ticket number {} is {}'.format(self.ticket_id, self.is_reserved))
            print('Venue and Timings are {} on {}'.format(self.venue, self.event_date))
            print('Your Seat No is {}'.format(self.seat_number))
            print('\n')
        
        else: 
            print('Your ticket number {} is {}'.format(self.ticket_id, self.is_reserved))
            print('\n')
            
            
######################################################################################################################################################################################################


B = TICKET('ABC123', 'music show', '01-01-2023', 'Town Hall', 'B-22', 2000, 'PENDING')
B.display_ticket_info()

print('AFTER USING RESERVE TICKET FUNCTION-->')
B.reserve_ticket('ABC123')
B.display_ticket_info()

print('AFTER USING CANCEL TICKET FUNCTION-->')
B.cancel_reservation('ABC123')
B.display_ticket_info()

print('IF WE ENTER A WRONG TICKET ID--')
B.reserve_ticket('EFG1234')

```

```python
# Q9. You are creating a shopping cart for an e-commerce website. Using OOP to model the "ShoppingCart" functionality the class should contain following attributes and methods:
#a. items: Represents the list of items in the shopping cart. 


#The class also includes the following methods:

#a. add_item(self, item): Adds an item to the shopping cart by appending it to the list of items.
#b. remove_item(self, item): Removes an item from the shopping cart if it exists in the list.
#c. view_cart(self): Displays the items currently present in the shopping cart.
#d. clear_cart(self): Clears all items from the shopping cart by reassigning an empty list to the items attribute.



class Shopping_cart:
    def __init__(self):
        self.item = []
    
    def add_item(self, item_1):
        if item_1 in self.item:
            print('Item already added in kart')
            
        else: self.item.append(item_1)
            
            
    def remove_item(self, item_1):
        if item_1 in self.item:
            self.item.remove(item_1)
        else:
            print('Item not available in kart')
            
    def clear_cart(self):
        self.item.clear()
        print('The kart is cleared')
        
        
######################################################################################################################################################################################################

KART = Shopping_cart()

print('Adding Item To The Kart')
KART.add_item('Mango')
KART.add_item('Phone')
KART.add_item('T-Shirt')
KART.add_item('Shorts')
KART.add_item('Bread')

print('Current Kart Content {} \n'.format(KART.item))

print('Adding Item From The Kart')
KART.remove_item('Mango')
KART.remove_item('Phone')
print('Current Kart Content {} \n'.format(KART.item))

print('Clearning the Kart')
KART.clear_cart()
print('Current Kart Content {} \n'.format(KART.item))
```
