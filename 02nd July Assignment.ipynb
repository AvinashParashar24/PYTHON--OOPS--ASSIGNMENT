```python
# Q1 Explain what inheritance is in object-oriented programming and why it is used.

# ANSWER -->
'''Inheritance allows us to define a class that inherits all the methods and properties from another class. Parent class is the class being inherited from, also called base class.
Child class is the class that inherits from another class, also called derived class.'''





################ PARENT CLASS ######################
class student:
    def __init__(self, name,age):
        self.name = name
        self.age= age
    def info_1(self):
        print('The stuednt is {} & his age is {}'.format(self.name, self.age))
        
################ CHILD CLASS ######################

        
class Avinash(student):
    def __init__(self, name,age, marks):

        super(). __init__(name,age) # WE ARE USING SUPER TO INHERIT NAME AND AGE FROM PARENT CLASS 
        self.marks = marks # WE ALSO HAVE A NEW CLASS WHICH IS NOT INHERTED HENCE WE ARE CREATING IT
    
    def info(self):
        super().info_1() # WE ARE INHERITING THE FUNCTIONS FROM PREVIOUSLY MADE CLASS "STUDENT"
        print('He scored {}'.format(self.marks))
        
################ OBJECT ######################

A = Avinash('Avin', 28,110)
A.info()
    
    
```

```python
# Q2 Discuss the concept of single inheritance and multiple inheritance, highlighting their differences and advantages.

# ANSWER -->
'''If a child class is inheriting the properties of a single other class, we call it single inheritance. However,
if a child class inherits from more than one class, i.e. this child class is derived from multiple classes, we call it multiple inheritance in Python.'''

# ADVANTAGES:-

'''1. Modular Codebase: Increases modularity, i.e., breaking down the codebase into modules, making it easier to understand. 
Here, each class we define becomes a separate module that can be inherited separately by one or many classes.'''

'''2. Code Reusability: the child class copies all the attributes and methods of the parent class into its class and use. 
It saves time and coding effort by not rewriting them, thus following modularity paradigms.'''

'''3. Less Development and Maintenance Costs: changes must be made in the base class, and all derived classes will automatically follow.'''

'''4. Reusability â€“ Inheritance allows obtaining new classes from existing classes without modification. This helps reusability of information in the child class and adds extra functionality.'''
```

```python
# Q4. What is the significance of the "protected" access modifier in inheritance? How does it differ from "private" and "public" modifiers?

#ANSWER 
'''The members of a class that are declared protected are only accessible to a class derived from it. 
Data members of a class are declared protected by adding a single underscore '_' symbol before the data member of that class.'''

'''This allows subclasses to access and potentially modify these members, which can be useful when you want to provide some level of controlled access to derived classes while still maintaining encapsulation.'''


''' In case of Private modifiers, the sub class cannot access any method or property from the parent class & they remain confined to the class in which they are defined.'''

''' In case of Public modifiers, any sub class can access the method and property from a parent class even if they are outside the hierarchy'''

# Summary --> 
'''In short, the Private modifiers cannnot be accessed by the sub class where as the public and protected can me accessed by the child class. 
The Protected modifiers provide additional security where it can be accessed by only its own derived class.'''


```

```python
# Q6. 
'''Create a base class called "Vehicle" with attributes like "make", "model", and "year". Then, create a derived class called "Car" that inherits from "Vehicle" and adds an attribute called "fuel_type".
Implement appropriate methods in both classes.'''


# ANSWER 


################ PARENT CLASS ####################
class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        
    def description (self):
        print('This is {} {} from {} year'.format(self.make, self.model, self.year))
    
        
A = Vehicle('Honda', 'Jazz', 2023)

print('Coming from Parent class-->')
A.description()

print('######################################\n')



################ CHILD CLASS ####################

class Car(Vehicle):
    def __init__(self, make, model, year, fuel_type):
        super().__init__(make, model, year)
        self.fuel_type = fuel_type
        
    def info(self):
        super().description()
        print('The fuel type of car is {}'.format(self.fuel_type))
        
        
B = Car('Toyota', 'Fortuner', 2020, 'Patrol')

print('Coming from Child class-->')
B.info()
        
        
    
```

```python
# Q7 

''''Create a base class called "Employee" with attributes like "name" and "salary." Derive two classes, "Manager" and "Developer," from "Employee." Add an additional
attribute called "department" for the "Manager" class and "programming_language" for the "Developer" class.'''

# ANSWER 

################ PARENT CLASS ####################

class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
        
    def Employee_info(self):
        print('The employee name is {} and his salary is {} INR'.format(self.name, self.salary))
        
avin = Employee('Avinash', 10000)

print('Coming from Parent class-->')
avin.Employee_info()
print('######################################\n')


################ CHILD CLASS 1  ####################

class Manager(Employee):
    def __init__(self, name, salary, department):
        super().__init__(name, salary)
        self.department = department
        
    def Manager_info(self):
        super().Employee_info()
        print('He is from {} Department'.format(self.department))
        
        
print('Coming from Child class 1-->')       
Man = Manager('Shak', 500000, 'GSCL')
Man.Manager_info()
print('######################################\n')

        
################ CHILD CLASS 2  ####################


class Developer(Employee):
    def __init__(self, name, salary, programming_language):
        super().__init__(name, salary)
        self.programming_language = programming_language
        
    def Developer_info(self):
        super().Employee_info()
        print('He is using {} Language'.format(self.programming_language))
        
        
print('Coming from Child class 2-->')       
Man = Developer('Vagish', 900000, 'Python')
Man.Developer_info()
print('######################################\n')
        


```

```python
# Q8. 
'''Design a base class called "Shape" with attributes like "colour" and "border_width." Create derived classes, "Rectangle" and "Circle," that inherit from "Shape" and add
specific attributes like "length" and "width" for the "Rectangle" class and "radius" for the "Circle" class.'''

# ANSWER ---> 




################ PARENT CLASS ####################

class Shape:
    def __init__(self, colour, border_width):
        self.colour = colour
        self.border_width = border_width
        
    def Shape_measurement(self):
        print('The shape has {} colour and its border width is {}'.format(self.colour, self.border_width))
        
GEOM = Shape('Red', '3.15mm')

print('Coming from Parent class-->')
GEOM.Shape_measurement()
print('######################################\n')


################ CHILD CLASS 1  ####################


class Rectangle(Shape):
    
    def __init__(self, colour, border_width,length, width):
        super().__init__(colour, border_width)
        self.length = length
        self.width = width
        
    def rect_measurement(self):
        super().Shape_measurement()
        print('The length of Rectangle is {} and the width is {}'.format(self.length, self.length))
        
        
print('Coming from Child class 1-->')       
Rect = Rectangle('Green','1.1mm', '2.5mm', '3mm')
Rect.rect_measurement()
print('######################################\n')


        
################ CHILD CLASS 2  ####################


class Circle(Shape):
    
    def __init__(self, colour, border_width, radius):
        super().__init__(colour, border_width)
        self.radius = radius
        
    def circle_measure(self):
        super().Shape_measurement()
        print('The radius of circle is {}'.format(self.radius))
        
print('Coming from Child class 2-->')    

Cir = Circle('black', '9mm', '5mm')
Cir.circle_measure()
print('######################################\n')
        
```
